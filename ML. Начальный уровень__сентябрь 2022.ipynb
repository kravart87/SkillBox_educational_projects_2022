&nbsp;&nbsp;&nbsp;&nbsp; Любой бизнес хочет максимизировать количество клиентов. Для достижения этой цели важно не только пытаться привлечь новых, но и удерживать уже существующих. Удержать клиента обойдется компании дешевле, чем привлечь нового. Кроме того, новый клиент может оказаться слабо заинтересованным в услугах бизнеса и с ним будет сложно работать, тогда как о старых клиентах уже есть необходимые данные по взаимодействию с сервисом. 
 <br>

&nbsp;&nbsp;&nbsp;&nbsp; Соответственно, прогнозируя отток, мы можем вовремя среагировать и попытаться удержать клиента, который хочет уйти. Опираясь на данные об услугах, которыми пользуется клиент, мы можем сделать ему специальное предложение, пытаясь изменить его решение об уходе от оператора. Благодаря этому задача удержания будет легче в реализации, чем задача привлечения новых пользователей, о которых мы еще ничего не знаем.<br>

&nbsp;&nbsp;&nbsp;&nbsp; Вам предоставлен набор данных от телекоммуникационной компании. В данных содержится информация о почти шести тысячах пользователей, их демографических характеристиках, услугах, которыми они пользуются, длительности пользования услугами оператора, методе оплаты, размере оплаты.

&nbsp;&nbsp;&nbsp;&nbsp; `PaperlessBilling` – пользуется ли клиент безбумажным биллингом (Yes, No)<br>
&nbsp;&nbsp;&nbsp;&nbsp; `PaymentMethod` – метод оплаты (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))<br>
&nbsp;&nbsp;&nbsp;&nbsp; `MonthlyCharges` – месячный размер оплаты на настоящий момент<br>
&nbsp;&nbsp;&nbsp;&nbsp; `TotalCharges` – общая сумма, которую клиент заплатил за услуги за все время<br>

```python
# Количество пользователей в разрезе различных категорий.

users_by_category = pd.DataFrame(columns = ['Категория', 'Кол-во пользователей'])

labels = [['Женщины', 'Мужчины'], ['Молодые', 'Пожилые'], ['Холостые/незамужние', 'Женатые/замужние'],
          ['Нет иждевенцев', 'Есть иждевенцы']]

for index, attribute in enumerate(['gender', 'SeniorCitizen', 'Partner', 'Dependents']):
  for value in [0, 1]:
      number_of_users = int(users_df[users_df[attribute] == value]['customerID'].count()) 
      users_by_category = users_by_category.append({'Категория': labels[index][value], 
                                            'Кол-во пользователей': number_of_users}, ignore_index=True)
      
plt.figure(figsize = (8, 5))
plt.title('Количество абонентов в разрезе категорий')
sns.barplot(x = 'Кол-во пользователей', y = 'Категория', data = users_by_category)
plt.ylabel(None)
plt.show()
```

***Далее, можно проверить, есть ли категории пользователей, которые явно выделяются среди других (например - женатые мужчины или молодые и холостые клиенты)***

```python
# Функция для визуализации количественного разделения пользователей по категориям.

def grouping_by_attributes(main_attribute, additional_attribute, title, labels, xticks):

  attribute_dataset = users_df.groupby([main_attribute, additional_attribute], as_index = False)['customerID']\
                              .count().rename(columns = {'customerID': 'Count'})
                              
  grid = sns.FacetGrid(attribute_dataset, col = additional_attribute, height = 4, aspect = 1.1)
  grid.map(plt.bar, main_attribute, 'Count', alpha=.5, color = ['green', 'blue'])
  grid.fig.suptitle(f'Распределение абонентов по признаку "{title}"', 
                    y = 1.05, fontsize = 12)
  grid.set_axis_labels(' ', 'Всего клиентов')
  for attr, ax in grid.axes_dict.items():
     ax.set_xticks(range(0, 2))
     ax.set_xticklabels([xticks[0], xticks[1]])
     if attr == 0:
          ax.set_title(labels[0], fontsize = 11)
     else:
          ax.set_title(labels[1], fontsize = 11)
  plt.show()
```

```python
# Пол.

title = ['Пол - Возраст', 'Пол - Семейное положение', 'Пол - Иждевенцы']
labels = [['Молодые', 'Пожилые'], ['Холостые/незамужние', 'Женатые/замужние'], 
          ['Нет иждевенцев', 'Есть иждевенцы']]
xticks = ['Женщины', 'Мужчины']
for index, additional_attribute in enumerate(['SeniorCitizen', 'Partner', 'Dependents']):
  grouping_by_attributes('gender', additional_attribute, title[index], labels[index], xticks)
  
```

*Как видно из приведенных выше данных, гендерная принадлежность не влияет на другие характеристики. То есть, среди мужчин и женщин оказалось равное количество пожилых, женатых/замужних или холостых/незамужних, а также имеющих или не имеющих иждевенцев.*

```python
# Семейное положение.

title = ['Семейное положение - Возраст', 'Семейное положение - Иждевенцы']
labels = [['Молодые', 'Пожилые'], ['Нет иждевенцев', 'Есть иждевенцы']]
xticks = ['Хол./незам.', 'В браке']

for index, additional_attribute in enumerate(['SeniorCitizen', 'Dependents']):
  grouping_by_attributes('Partner', additional_attribute, title[index], labels[index], xticks)
```

*Судя по графику, видно равномерное распределение внутри пожилых и молодых людей по семейному положению. То есть, можно сказать, что как среди пенсионеров, так и среди молодых людей, соотношение семейных и холостых/незамужних составляет 50/50.*\
*Среди холостых/незамужних абонентов только 10% имеют на своем содержании иждевенцев. Среди семейных пар половина пользователей имеет иждевенцев.*\
*Другими словами, 80% тех кто имеет иждевенцев - это женатые/замужние.*

```python
# Возраст и иждевенцы.

title = 'Возраст - Иждевенцы'
labels = ['Молодые', 'Пожилые']
xticks = ['Нет ижд-цев', 'Есть ижд-цы']
grouping_by_attributes('Dependents', 'SeniorCitizen', title, labels, xticks)
```

*Практически все абоненты, у которых есть иждевенцы - это молодые люди. Среди пенсионеров, доля таких пользователей не превышает 5%.*

```python
# Продолжительность пользования услугами оператора в зависимости от категории клиента.

users_tenure = pd.DataFrame(columns = ['Категория', 't Обслуживания'])

for index, attribute in enumerate(['gender', 'SeniorCitizen', 'Partner', 'Dependents']):
  for value in [0, 1]:
      tenure_df = users_df.groupby([attribute], as_index = False)['tenure'].mean().round(1)
      tenure = tenure_df[tenure_df[attribute] == value]['tenure'].values[0]
      users_tenure = users_tenure.append({'Категория': labels[index][value],
                                        't Обслуживания': tenure}, ignore_index=True)

plt.figure(figsize = (8, 5))
sns.barplot(x = 't Обслуживания', y = 'Категория', data = users_tenure)
plt.title('Медиана продолжительности обслуживания по категориям пользователей', y = 1.04, fontsize = 12)
plt.ylabel(None)
plt.xticks(range(0, 50, 5))
plt.axvline(users_df['tenure'].mean(), linestyle = '--', color = 'black')
plt.show()
```

*В среднем, дольше всех являются клиентами оператора состоящие в браке и имеющие иждевенцев  - 42 и 38 месяцев. Самый низкий показатель среди абонентов, не состоящих в браке - 23 месяца.*

# 4. Обработка данных и построение модели

Для данного набора данных я буду использовать два типа моделей - линейные и на основе деревьев решений. В силу различий этих алгоритмов и процесс подготовки признаков будет различаться.

```python
prepared_data = users_df.copy()
```

*В соответствии с выявленными ранее зависимостями, проведу преобразования данных.*

```python
# Новый признак 'женатые/замужние абоненты с детьми'.
prepared_data['Married&Children'] = 0
prepared_data.loc[(prepared_data['Partner'] == 1) & (prepared_data['Dependents'] == 1), 'Married&Children'] = 1

# Т.к. были выявлены связи между подключенными услугами и оттоком, введу два признака:
# Support&Security - если у абонента подключена услуга OnlineSecurity или TechSupport.
# Streaming - если у абонента подключены стриминговые сервисы ТВ или кино.
prepared_data['Support&Security'] = 0
prepared_data.loc[(prepared_data['OnlineSecurity'] == 1) | (prepared_data['TechSupport'] == 1), 'Support&Security'] = 1
prepared_data['Streaming'] = 0
prepared_data.loc[(prepared_data['StreamingTV'] == 1) | (prepared_data['StreamingMovies'] == 1), 'Streaming'] = 1

prepared_data['MultipleLines'] = prepared_data['MultipleLines'].map({'Yes': 1, 'No': 0, 'No phone service': 0})

# Удаление лишних признаков.
prepared_data = prepared_data.drop(['Unnamed: 0', 'customerID', 'gender', 'Partner', 'Dependents'], axis = 1)
prepared_data = prepared_data.drop(additional_services, axis = 1)

# Кодироване признаков с помощью метода get_dummies.
transform_set = pd.get_dummies(prepared_data)
features = transform_set.drop(['Churn'], axis = 1)
targets = transform_set['Churn']
```

```python
transform_set.corr()['Churn'].sort_values()[:-1].plot(kind = 'bar', figsize = (10, 6), ylabel = 'к-т корреляции', 
                                                      title = 'Корреляция отобранных признаков с целевым зачением' )
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_auc_score, f1_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, targets, test_size = 0.2, 
                                                    stratify = targets, random_state = 17)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

#### **KNeighborsClassifier**

```python
knn = KNeighborsClassifier(n_neighbors = 35)
knn_model = knn.fit(x_train_scaled, y_train)
knn_predict = knn_model.predict(x_test_scaled)
knn_predict_proba = knn_model.predict_proba(x_test_scaled)[:, 1]

print('KNeighborsClassifier')
print('*' * 35)
print(f'Правильность на обучающем наборе: {round(knn_model.score(x_train_scaled, y_train), 4)}')
print()
print(f'Правильность на тестовом наборе: {round(knn_model.score(x_test_scaled, y_test), 4)}')
print(f'f1_score: {round(f1_score(y_test, knn_predict)*100, 4)}')
print(f'ROC_AUC_proba: {round(roc_auc_score(y_test, knn_predict_proba)*100, 4)}')
```

#### **LogisticRegression**

```python
lrg = LogisticRegression(solver = 'liblinear', penalty = 'l2', C = 0.8, random_state = 17, 
                         class_weight = {1: 1.4}, max_iter = 1000)
lrg_model = lrg.fit(x_train_scaled, y_train)
lrg_predict = lrg_model.predict(x_test_scaled)
lrg_predict_proba = lrg_model.predict_proba(x_test_scaled)[:, 1]

print('LogisticRegression')
print('*' * 35)
print(f'Правильность на обучающем наборе: {round(lrg_model.score(x_train_scaled, y_train), 4)}')
print()
print(f'Правильность на тестовом наборе: {round(lrg_model.score(x_test_scaled, y_test), 4)}')
print(f'f1_score: {round(f1_score(y_test, lrg_predict)*100, 4)}')
print(f'ROC_AUC: {round(roc_auc_score(y_test, lrg_predict)*100, 4)}')
print(f'ROC_AUC_proba: {round(roc_auc_score(y_test, lrg_predict_proba)*100, 4)}')
```

#### **SVC**

```python
svc = SVC(C = 0.5, class_weight = {1: 1.2}, probability = True, random_state = 42)
svc_model = svc.fit(x_train_scaled, y_train)
svc_predict = svc_model.predict(x_test_scaled)
svc_predict_proba = svc_model.predict_proba(x_test_scaled)[:, 1]

print('SVC')
print('*' * 35)
print(f'Правильность на обучающем наборе: {round(svc_model.score(x_train_scaled, y_train), 4)}')
print()
print(f'Правильность на тестовом наборе: {round(svc_model.score(x_test_scaled, y_test), 4)}')
print(f'f1_score: {round(f1_score(y_test, svc_predict)*100, 4)}')
print(f'ROC_AUC: {round(roc_auc_score(y_test, svc_predict)*100, 4)}')
print(f'ROC_AUC_proba: {round(roc_auc_score(y_test, svc_predict_proba)*100, 4)}')
```

#### **RandomForestClassifier**

```python
rfc = RandomForestClassifier(n_estimators = 500, max_depth = 12, min_samples_split = 3,
                             class_weight = {1: 1.4}, min_weight_fraction_leaf = 0.002, random_state = 17)
rfc_model = rfc.fit(x_train_scaled, y_train)
rfc_predict = rfc_model.predict(x_test_scaled)
rfc_predict_proba = rfc_model.predict_proba(x_test_scaled)[:, 1]

print('RandomForestClassifier')
print('*' * 35)
print(f'Правильность на обучающем наборе: {round(rfc_model.score(x_train_scaled, y_train), 4)}')
print()
print(f'Правильность на тестовом наборе: {round(rfc_model.score(x_test_scaled, y_test), 4)}')
print(f'f1_score: {round(f1_score(y_test, rfc_predict)*100, 4)}')
print(f'ROC_AUC: {round(roc_auc_score(y_test, rfc_predict)*100, 4)}')
print(f'ROC_AUC_proba: {round(roc_auc_score(y_test, rfc_predict_proba)*100, 4)}')
```

#### **GradientBoostingClassifier**

# Итоговый выбор модели

```python
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve
```

*Из всех представленных моделей более оптимальные результаты показал **GradientBoostingClassifier**. Но, в то же самое время, если рассмативать с точки зрения метрики f1_score, то предочтительнее выглядит **RandomForestClassifier**.*

*Необходимо более подробно рассмотреть данные алгоритмы.*

```python
# Визуализация матрицы ошибок

def plot_confusion_matrix_illustration(conf_matrix, model_name):
    plt.figure(figsize = (6, 6))
    confusion = conf_matrix
    plt.title(f'Матрица ошибок для {model_name}', fontsize = 18, y = 1.05)
    plt.text(0.40, 0.7, confusion[0, 0], size = 50, horizontalalignment = 'right')
    plt.text(0.40, 0.2, confusion[1, 0], size = 50, horizontalalignment = 'right')
    plt.text(0.90, 0.7, confusion[0, 1], size = 50, horizontalalignment = 'right')
    plt.text(0.90, 0.2, confusion[1, 1], size = 50, horizontalalignment = 'right')
    plt.xticks([0.25, 0.75], ['predicted "not churn"', 'predicted "churn"'], size = 15)
    plt.yticks([0.25, 0.75], ['true "churn"', 'true "not churn"'], size = 15)
    plt.plot([0.5, 0.5], [0, 1], '--', c = 'k')
    plt.plot([0, 1], [0.5, 0.5], '--', c = 'k')

    plt.xlim(0, 1)
    plt.ylim(0, 1)
```

```python
plot_confusion_matrix_illustration(confusion_matrix(y_test, rfc_predict), 'RandomForestClassifier')
```

```python
plot_confusion_matrix_illustration(confusion_matrix(y_test, gbc_predict), 'GradientBoostingClassifier')
```

***C одной стороны, мы видим, что RandomForestClassifier лучше предсказывает пользователей, которые приняли решение об уходе, но и допускает больше ошибок, классифицируя лояльных клиентов как ушедших. У алгоритма GradientBoostingClassifier ситуация обратная.***

***Чаще всего, в задачах такого характера, предпочтение отдается в пользу более точного алгоритма, но как видно было ранее, пользователи, которые приняли решение об уходе, приносят больше прибыли, поэтому выявить больше таких абонентов может быть предпочтительнее.***

```python
plt.figure(figsize = (9, 5))
plt.title('Сравнение кривых точности-полноты RandomForestClassifier и GradientBoostingClassifier')

gbc_fpr, gbc_tpr, gbc_thresholds = precision_recall_curve(y_test, gbc_predict_proba)
rfc_fpr, rfc_tpr, rfc_thresholds = precision_recall_curve(y_test, rfc_predict_proba)
plt.plot(gbc_fpr, gbc_tpr, label = 'кривая точности-полноты GBC')
plt.plot(rfc_fpr, rfc_tpr, label = 'кривая точности-полноты RFC')
plt.xlabel('Точность')
plt.ylabel('Полнота')
plt.grid(color = 'gray', linestyle = '--', alpha  = 0.8)
plt.legend(loc = 'best')

plt.figure(figsize = (9, 5))
plt.title('Сравнение ROC кривых для RandomForestClassifier и GradientBoostingClassifier')
gbc_fpr, gbc_tpr, gbc_thresholds = roc_curve(y_test, gbc_predict_proba)
rfc_fpr, rfc_tpr, rfc_thresholds = roc_curve(y_test, rfc_predict_proba)
plt.plot(gbc_fpr, gbc_tpr, label = 'ROC-кривая GBC')
plt.plot(rfc_fpr, rfc_tpr, label = 'ROC-кривая RFC')
plt.xlabel('FPR')
plt.ylabel('TPR (полнота)')
plt.grid(color = 'gray', linestyle = '--', alpha  = 0.8)
plt.legend(loc = 4)
plt.show()
```

**В качестве итоговой модели быдет выбран GradientBoostingClassifier.**\
**Исходя из представленных графиков, можно сделать вывод, что данный алгоритм способен показать то же значение полноты, но без потери точности.**
